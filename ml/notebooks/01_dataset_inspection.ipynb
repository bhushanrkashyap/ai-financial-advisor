# 01 — SCF 2022 public dataset inspection

Exploratory look at **rscfp2022.dta** (Summary Extract for the 2022 Survey of Consumer Finances). No cleaning or feature engineering — inspection only.

Expected path (relative to repository root): `datasets/raw/scfp2022s/rscfp2022.dta`

```python
from pathlib import Path

import pandas as pd


def find_repo_root() -> Path:
    """Resolve monorepo root whether Jupyter cwd is project root or ml/notebooks."""
    cwd = Path.cwd().resolve()
    for directory in (cwd, *cwd.parents):
        if (directory / "backend").is_dir() and (directory / "ml").is_dir():
            return directory
    return cwd


REPO_ROOT = find_repo_root()
DATA_PATH = REPO_ROOT / "datasets" / "raw" / "scfp2022s" / "rscfp2022.dta"

if not DATA_PATH.is_file():
    raise FileNotFoundError(
        f"SCF extract not found at {DATA_PATH}. "
        "Place rscfp2022.dta under datasets/raw/scfp2022s/ (see Federal Reserve SCF)."
    )

df = pd.read_stata(DATA_PATH)
print(f"Loaded: {DATA_PATH}")
```

## Missing values summary

```python
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(2)
summary = (
    pd.DataFrame({"missing_count": missing, "missing_pct": missing_pct})
    .sort_values("missing_count", ascending=False)
)
print(
    "Columns with any missing:",
    int((summary["missing_count"] > 0).sum()),
    "of",
    len(summary),
)
summary
```
