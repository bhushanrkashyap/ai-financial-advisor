# 03 — Lending Club (accepted loans) inspection

Exploratory slice of the **accepted loans** CSV under `datasets/raw/lendingclub/`. Only the **first 50,000 rows** are read (`nrows=50000`). No cleaning or feature engineering.

Default file: `accepted_2007_to_2018Q4.csv` (override `CSV_NAME` in the load cell if your extract uses a different name).

```python
from pathlib import Path

import pandas as pd

NROWS = 50_000
CSV_NAME = "accepted_2007_to_2018Q4.csv"


def find_repo_root() -> Path:
    cwd = Path.cwd().resolve()
    for directory in (cwd, *cwd.parents):
        if (directory / "backend").is_dir() and (directory / "ml").is_dir():
            return directory
    return cwd


REPO_ROOT = find_repo_root()
LC_DIR = REPO_ROOT / "datasets" / "raw" / "lendingclub"
DATA_PATH = LC_DIR / CSV_NAME

if not DATA_PATH.is_file():
    raise FileNotFoundError(
        f"Lending Club CSV not found at {DATA_PATH}. "
        f"Place the accepted-loans export in {LC_DIR} or set CSV_NAME."
    )

df = pd.read_csv(DATA_PATH, nrows=NROWS, low_memory=False)
print(f"Loaded: {DATA_PATH}  |  nrows cap: {NROWS}  |  shape: {df.shape}")
```

## Missing values summary

```python
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(2)
summary = (
    pd.DataFrame({"missing_count": missing, "missing_pct": missing_pct})
    .sort_values("missing_count", ascending=False)
)
print(
    "Columns with any missing:",
    int((summary["missing_count"] > 0).sum()),
    "of",
    len(summary),
)
summary
```
